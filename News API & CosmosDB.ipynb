sample query (loop bedingung müsste eigenschaft "totalResults" aus erstem query ergebnis sein)

```python
companies=["Facebook","Google","Amazon","Netflix","Apple"]
my_urls=[]
articles={}
m=0
#automate datetime pl0x
for i in range(len(companies)):
    articles[companies[i]] = newsapi.get_everything(q=companies[i], 
                                      sources='google news', 
                                      from_param='2018-07-23', 
                                      to='2018-07-29', 
                                      page_size=100,
                                      page=1)
    print(articles[companies[i]])
    m+=1
    
```

extract individual articles

```python
for i in range(len(companies)):
    for j in articles[companies[i]]["articles"]:
        print(j["url"])
        my_urls.append(j["url"])
```

```python
#Built a scrapy spider
```

```python
for url in my_urls:
    source_code=requests.get(url)
    plain_text=source_code.text
    soup_object=soup(plain_text,"xml")
    for text in soup_object.findAll("p"):
        print(text)
        print("-"*60)
```
